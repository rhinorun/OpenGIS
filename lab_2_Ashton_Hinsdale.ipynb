```python
import pandas as pd
import json
```

```python
data = json.load(open('fsq_coffee.json'))
df = pd.DataFrame(data["results"])
```

```python
df
```

```python
#I'm gonna extract the lats and longs brother
df['latitude'] = df['geocodes'].apply(lambda x: x['main']['latitude'])
df['longitude'] = df['geocodes'].apply(lambda x: x['main']['longitude'])
```

```python
df
```

```python
import folium
```

```python
#creating a base map brother
center_lat = df['latitude'].mean()
center_lng = df['longitude'].mean()
covfefe_map = folium.Map(location=[center_lat, center_lng], zoom_start=18)

#adding markers for every point in tha dataframe brotherrrrrrrr
for idx, row in df.iterrows():
    folium.Marker(
        location=[row['latitude'], row['longitude']],
        popup=row['name']
    ).add_to(covfefe_map)

covfefe_map
```
